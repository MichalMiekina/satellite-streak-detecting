```python
import pandas as pd

# Load the CSV file
# data = pd.read_csv('model_accuracies_resnet_2024-04-17_14-20-29.csv')
# data = pd.read_csv('model_metrics_resnet_2024-04-17_16-58-39.csv')
# data = pd.read_csv('model_accuracies_mobilenetv2_2024-04-17_16-30-09.csv')
# data = pd.read_csv('model_metrics_mobilenetv2_2024-04-17_17-23-06.csv')
data = pd.read_csv('model_metrics_resnet_2024-05-09_16-02-22.csv')

# Calculate statistics for numeric columns
statistics = data.describe()

# Display results
print("Statistics for Numeric Columns:\n", statistics)

# Calculate mean, standard deviation, median, min, and max for accuracy, precision, recall, and F1-score
accuracy_mean = data['Accuracy'].mean()
accuracy_std_dev = data['Accuracy'].std()
accuracy_median = data['Accuracy'].median()
accuracy_min = data['Accuracy'].min()
accuracy_max = data['Accuracy'].max()

precision_mean = data['Precision'].mean()
precision_std_dev = data['Precision'].std()
precision_median = data['Precision'].median()
precision_min = data['Precision'].min()
precision_max = data['Precision'].max()

recall_mean = data['Recall'].mean()
recall_std_dev = data['Recall'].std()
recall_median = data['Recall'].median()
recall_min = data['Recall'].min()
recall_max = data['Recall'].max()

f1_mean = data['F1'].mean()
f1_std_dev = data['F1'].std()
f1_median = data['F1'].median()
f1_min = data['F1'].min()
f1_max = data['F1'].max()

# Display results for accuracy
print("\nAccuracy Mean:", accuracy_mean)
print("Accuracy Standard Deviation:", accuracy_std_dev)
print("Accuracy Median:", accuracy_median)
print("Accuracy Minimum:", accuracy_min)
print("Accuracy Maximum:", accuracy_max)

# Display results for precision
print("\nPrecision Mean:", precision_mean)
print("Precision Standard Deviation:", precision_std_dev)
print("Precision Median:", precision_median)
print("Precision Minimum:", precision_min)
print("Precision Maximum:", precision_max)

# Display results for recall
print("\nRecall Mean:", recall_mean)
print("Recall Standard Deviation:", recall_std_dev)
print("Recall Median:", recall_median)
print("Recall Minimum:", recall_min)
print("Recall Maximum:", recall_max)

# Display results for F1-score
print("\nF1-score Mean:", f1_mean)
print("F1-score Standard Deviation:", f1_std_dev)
print("F1-score Median:", f1_median)
print("F1-score Minimum:", f1_min)
print("F1-score Maximum:", f1_max)
```
